```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_df = sns.load_dataset('titanic')
```

```python
# Conta nulos em cada linha
titanic_df["nulos_por_linha"] = titanic_df.isnull().sum(axis=1)

# Visualizar as primeiras linhas
print(titanic_df[["nulos_por_linha"]].head())
```

```python
titanic_df["nulos_por_linha"] = titanic_df.isnull().sum(axis=1)

linhas_com_nulos = titanic_df[titanic_df["nulos_por_linha"] > 0]

# Plotando a distribuição dos nulos por linha
plt.figure(figsize=(6,4))
linhas_com_nulos["nulos_por_linha"].value_counts().sort_index().plot(kind="bar")
plt.xlabel("Quantidade de nulos por linha")
plt.ylabel("Número de linhas")
plt.title("Distribuição de valores nulos por linha")
plt.show()
```
