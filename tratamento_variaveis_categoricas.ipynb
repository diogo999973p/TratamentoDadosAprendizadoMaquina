```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

titanic_df = sns.load_dataset("titanic")

columns_to_drop = ['embarked', 'embark_town', 'alive', 'deck']

titanic_df = titanic_df.drop(columns=columns_to_drop)

titanic_encoded_df = pd.get_dummies(titanic_df, columns=["sex", "who"], drop_first=True)

titanic_encoded_df["class_encoded"] = le.fit_transform(titanic_encoded_df["class"])

titanic_encoded_df = titanic_encoded_df.drop(columns=['class'])
```
